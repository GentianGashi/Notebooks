## Description : 
Below is an attempt to visually summarize the results of a <a href="https://cocl.us/datascience_survey_data">survey</a> that was conducted to gauge an audience interest in different data science topics. Scroll to the end of the document for the graphs and references.

```python
import numpy as np  # (Arrays & Matrices)
import pandas as pd # (Data Structures & Tools)
```

```python
# Loading CSV file into dataframe:
survey_df = pd.read_csv('https://cocl.us/datascience_survey_data', index_col = 0)
survey_df.head()
```

```python
# Sorting dataframe by the 'Very Interested' column in descending order:
survey_df.sort_values(by='Very interested', ascending=False, inplace = True)
survey_df.head()
```

```python
%matplotlib inline 

# Importing Matplotlib library
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.patches as patches

mpl.style.use('ggplot') # optional: changes style of chart

# line to plot percentages instead of absolute values
ax = (survey_df.div(survey_df.sum(1), axis=0)).plot(kind='bar', figsize=(20, 8), width = 0.8, color = ['#5cb85c','#5bc0de','#d9534f'])
plt.legend(labels=survey_df.columns,fontsize= 14)
plt.title("Percentage of Respondents' Interest in Data Science Areas",fontsize= 16)

plt.xticks(fontsize=14)
for spine in plt.gca().spines.values():
    spine.set_visible(False)
plt.yticks([])

# loop to add annotations
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy() 
    ax.annotate('{:.0%}'.format(height), (x, y + height + 0.01))
# Shows barchart
plt.show()
```
